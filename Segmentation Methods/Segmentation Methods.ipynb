```python
import torch
import numpy as np
import cv2
from PIL import Image
from transformers import GroundingDinoProcessor, GroundingDinoForObjectDetection
from ultralytics import SAM
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Load GroundingDINO
processor = GroundingDinoProcessor.from_pretrained("IDEA-Research/grounding-dino-base")
dino_model = GroundingDinoForObjectDetection.from_pretrained("IDEA-Research/grounding-dino-base")
dino_model.to(device).eval()

# Load SAM
sam_model = SAM("sam2_l.pt")
sam_model.model.to(device)
print("SAM model device:", next(sam_model.model.parameters()).device)

def preprocess_caption(caption: str) -> str:
    caption = caption.lower().strip()
    return caption if caption.endswith('.') else caption + '.'

def predict_single_image(image_path, processor, dino_model, sam_model, device):
    image = Image.open(image_path).convert("RGB")
    image_rgb = np.array(image)
    H, W = image_rgb.shape[:2]

    # GroundingDINO detection
    inputs = processor(images=image, text=preprocess_caption("human eye"), return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = dino_model(**inputs)

    results = processor.image_processor.post_process_object_detection(
        outputs, target_sizes=[(H, W)], threshold=0.4
    )[0]
    boxes = results["boxes"]
    scores = results["scores"]
    mask = (scores > 0.2) & (scores < 0.9)
    filtered_boxes = boxes[mask].cpu().numpy()

    if len(filtered_boxes) == 0:
        print("No eyes detected.")
        return None

    # ---- 1. Show GroundingDINO result ----
    dino_img = image_rgb.copy()
    for box in filtered_boxes:
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(dino_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    plt.figure(figsize=(6, 6))
    plt.imshow(dino_img)
    plt.title("GroundingDINO Detected Boxes")
    plt.axis("off")
    plt.show()

    # ---- 2. Run SAM and overlay all masks on one image ----
    sam_results = sam_model(image_path, bboxes=filtered_boxes.tolist(), verbose=False)

    overlay = image_rgb.copy()
    valid_count = 0

    # Define strong distinct colors (BGR format for OpenCV)
    strong_colors = [
        (255, 0, 0),     # Red
        (0, 255, 0),     # Green
        (0, 0, 255),     # Blue
        (255, 255, 0),   # Yellow
        (255, 0, 255),   # Magenta
        (0, 255, 255),   # Cyan
        (255, 128, 0),   # Orange
        (128, 0, 255)    # Purple
    ]

    for i, res in enumerate(sam_results):
        masks = res.masks.data.cpu().numpy().astype(np.uint8)

        for j, mask in enumerate(masks):
            if np.count_nonzero(mask) / (H * W) > 0.95:
                continue

            color = strong_colors[valid_count % len(strong_colors)]
            # Fill mask with strong color
            overlay[mask == 1] = (
                0.5 * overlay[mask == 1] + 0.5 * np.array(color)
            ).astype(np.uint8)

            valid_count += 1

    plt.figure(figsize=(6, 6))
    plt.imshow(overlay)
    plt.title(f"SAM Segmentations (Total = {valid_count})")
    plt.axis("off")
    plt.show()


    if valid_count == 0:
        print("No valid segmented region.")
    else:
        print(f"Total valid segmentations processed: {valid_count}")


# Example call
predict_single_image("/content/sample_data/50.jpg", processor, dino_model, sam_model, device)
```

```python
import os
import time
import torch
import numpy as np
from PIL import Image
import cv2
from transformers import GroundingDinoProcessor, GroundingDinoForObjectDetection
from ultralytics import SAM
from tqdm.notebook import tqdm

# === Device Setup ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# === Load Models ===
processor = GroundingDinoProcessor.from_pretrained("IDEA-Research/grounding-dino-base")
dino_model = GroundingDinoForObjectDetection.from_pretrained("IDEA-Research/grounding-dino-base")
dino_model.to(device)
dino_model.eval()

sam_model = SAM("sam2_l.pt")
sam_model.model.to(device)
print(f"SAM model running on: {next(sam_model.model.parameters()).device}")

# === Caption Preprocessing ===
def preprocess_caption(caption: str) -> str:
    result = caption.lower().strip()
    return result if result.endswith(".") else result + "."

# === Single Image Processing ===
def process_image(image_path, current_output_dir):
    filename = os.path.basename(image_path)
    try:
        image = Image.open(image_path).convert("RGB")
        image_rgb = np.array(image)
        H, W = image_rgb.shape[:2]

        # Grounding DINO inference
        inputs = processor(images=image, text=preprocess_caption("human eye"), return_tensors="pt")
        inputs = {k: v.to(device) for k, v in inputs.items()}
        with torch.no_grad():
            outputs = dino_model(**inputs)

        postprocessed_outputs = processor.image_processor.post_process_object_detection(
            outputs, target_sizes=[(H, W)], threshold=0.4
        )
        results = postprocessed_outputs[0]
        boxes = results["boxes"]
        scores = results["scores"]
        mask = (scores > 0.2) & (scores < 0.8)
        filtered_boxes = boxes[mask].cpu().numpy()

        if len(filtered_boxes) == 0:
            return f"No eye detected in {filename}"

        # SAM segmentation
        sam_results = sam_model(image_path, bboxes=filtered_boxes.tolist(), verbose=False)

        saved_count = 0
        for i, res in enumerate(sam_results):
            masks = res.masks.data.cpu().numpy().astype(np.uint8)
            for j, mask in enumerate(masks):
                nonzero_pixels = np.count_nonzero(mask)
                pixel_ratio = nonzero_pixels / (H * W)
                if pixel_ratio > 0.95:
                    # Skip masks with too many pixels
                    continue

                mask_3ch = np.repeat(mask[:, :, np.newaxis], 3, axis=2)
                extracted = image_rgb * mask_3ch
                x, y, w, h = cv2.boundingRect(mask)
                cropped = extracted[y:y+h, x:x+w]
                cropped_original = image_rgb[y:y+h, x:x+w]

                if cropped.shape != cropped_original.shape:
                    continue

                diff = np.abs(cropped.astype(np.float32) - cropped_original.astype(np.float32))
                mean_diff = np.mean(diff)
                if mean_diff < 1:
                    # Skip masks too similar to the original image
                    continue

                output_filename = f"{os.path.splitext(filename)[0]}_res{i}_mask{j}.png"
                output_path = os.path.join(current_output_dir, output_filename)
                cv2.imwrite(output_path, cv2.cvtColor(cropped, cv2.COLOR_RGB2BGR))
                saved_count += 1

        if saved_count == 0:
            return f"No valid masks saved for {filename}"
        return f"Processed {filename}"

    except Exception as e:
        return f"Error processing {filename}: {e}"

# === Directory Processing Function ===
def process_directory(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    image_files = [f for f in os.listdir(input_dir) if f.lower().endswith(".jpg")]
    total_images = len(image_files)

    start_time = time.time()
    results = []

    for fname in tqdm(image_files, desc=f"Processing {os.path.basename(input_dir)}"):
        image_path = os.path.join(input_dir, fname)
        result = process_image(image_path, output_dir)
        results.append(result)

    end_time = time.time()
    elapsed = end_time - start_time

    print(f"Finished {input_dir} in {elapsed:.2f} sec (avg {elapsed/total_images:.2f} sec/image)")

# === Process all subdirectories for train, val, and test ===
spl_data_dir = "/content/spl_data"  # Path to the main folder containing train, val, and test folders
spl_data_extracted_eye_dir = "/content/spl_data_extracted_eye"  # Path to the main folder containing train, val, and test folders
categories = ['50', '50-100', '100']

for split in ['train', 'val', 'test']:
    for category in categories:
        input_dir = os.path.join(spl_data_dir, split, category)
        output_dir = os.path.join(spl_data_extracted_eye_dir, split, f"extracted_eyes_{category}")
        os.makedirs(output_dir, exist_ok=True)  # Ensure the output directory exists
        process_directory(input_dir, output_dir)

print("Dataset processing completed!")
```

```python
from google.colab import drive
import os
import numpy as np
import cv2
from skimage.feature import hog, local_binary_pattern
from skimage.color import rgb2gray
from skimage.io import imread
from tqdm import tqdm

# Mount Google Drive
drive.mount('/content/drive')

# Define paths
data_dir = "/content/spl_data_extracted_eye"  # Path to the dataset
categories = ['100', '50', '50-100']  # Categories

# Define paths for saving features
drive_save_dir = "/content/drive/MyDrive"  # Path to save features in Google Drive

# Create directory in Google Drive if it doesn't exist
os.makedirs(drive_save_dir, exist_ok=True)

# Feature extraction parameters
radius = 1
n_points = 8 * radius
hog_params = {
    'orientations': 9,
    'pixels_per_cell': (8, 8),
    'cells_per_block': (2, 2),
    'block_norm': 'L2-Hys'
}

# Initialize lists for feature extraction
X_train = []
y_train = []
X_val = []
y_val = []

# === Processing the training set ===
for category in categories:
    class_dir = os.path.join(data_dir, 'train', f"extracted_eyes_{category}")
    for fname in tqdm(os.listdir(class_dir), desc=f"Processing {category} in train"):
        img_path = os.path.join(class_dir, fname)
        img = imread(img_path)
        if img is None or len(img.shape) != 3:
            continue  # Skip corrupted or grayscale images

        # Resize & convert to grayscale
        img = cv2.resize(img, (128, 128))
        gray_float = rgb2gray(img)
        gray_uint8 = (gray_float * 255).astype(np.uint8)

        # === Extract HOG features (multi-scale) ===
        hog_feat = []
        for scale in [0.5, 1, 1.5]:  # Multi-scale
            scaled_img = cv2.resize(gray_float, (int(128 * scale), int(128 * scale)))
            hog_feat.append(hog(scaled_img, **hog_params))
        hog_feat = np.concatenate(hog_feat)

        # === Extract LBP histogram features ===
        lbp = local_binary_pattern(gray_uint8, n_points, radius, method='uniform')
        (hist, _) = np.histogram(lbp.ravel(),
                                 bins=np.arange(0, n_points + 3),
                                 range=(0, n_points + 2))
        hist = hist.astype("float")
        hist /= (hist.sum() + 1e-6)

        # === Local brightness analysis ===
        local_mean = np.mean(gray_uint8)
        local_variance = np.var(gray_uint8)

        # === Concatenate features ===
        final_feat = np.concatenate([
            hog_feat,
            hist,
            [local_mean, local_variance]
        ])

        X_train.append(final_feat)
        y_train.append(category)

# === Processing the validation set ===
for category in categories:
    class_dir = os.path.join(data_dir, 'val', f"extracted_eyes_{category}")
    for fname in tqdm(os.listdir(class_dir), desc=f"Processing {category} in val"):
        img_path = os.path.join(class_dir, fname)
        img = imread(img_path)
        if img is None or len(img.shape) != 3:
            continue  # Skip corrupted or grayscale images

        # Resize & convert to grayscale
        img = cv2.resize(img, (128, 128))
        gray_float = rgb2gray(img)
        gray_uint8 = (gray_float * 255).astype(np.uint8)

        # === Extract HOG features (multi-scale) ===
        hog_feat = []
        for scale in [0.5, 1, 1.5]:  # Multi-scale
            scaled_img = cv2.resize(gray_float, (int(128 * scale), int(128 * scale)))
            hog_feat.append(hog(scaled_img, **hog_params))
        hog_feat = np.concatenate(hog_feat)

        # === Extract LBP histogram features ===
        lbp = local_binary_pattern(gray_uint8, n_points, radius, method='uniform')
        (hist, _) = np.histogram(lbp.ravel(),
                                 bins=np.arange(0, n_points + 3),
                                 range=(0, n_points + 2))
        hist = hist.astype("float")
        hist /= (hist.sum() + 1e-6)

        # === Local brightness analysis ===
        local_mean = np.mean(gray_uint8)
        local_variance = np.var(gray_uint8)

        # === Concatenate features ===
        final_feat = np.concatenate([
            hog_feat,
            hist,
            [local_mean, local_variance]
        ])

        X_val.append(final_feat)
        y_val.append(category)

# Convert lists to numpy arrays
X_train = np.array(X_train)
y_train = np.array(y_train)
X_val = np.array(X_val)
y_val = np.array(y_val)

print("Training set size:", X_train.shape[0], " Feature dimension:", X_train.shape[1])
print("Validation set size:", X_val.shape[0], " Feature dimension:", X_val.shape[1])

# === Save features to Google Drive ===
np.save(os.path.join(drive_save_dir, "X_train.npy"), X_train)
np.save(os.path.join(drive_save_dir, "y_train.npy"), y_train)
np.save(os.path.join(drive_save_dir, "X_val.npy"), X_val)
np.save(os.path.join(drive_save_dir, "y_val.npy"), y_val)

print("Features saved to Google Drive.")
```

```python
from google.colab import drive
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# Step 1: Mount Google Drive
drive.mount('/content/drive')

# Step 2: Load features from Google Drive
X_train = np.load('/content/drive/MyDrive/X_train.npy')
y_train = np.load('/content/drive/MyDrive/y_train.npy')
X_val = np.load('/content/drive/MyDrive/X_val.npy')
y_val = np.load('/content/drive/MyDrive/y_val.npy')

print("Training set size:", X_train.shape[0], " Feature dimension:", X_train.shape[1])
print("Validation set size:", X_val.shape[0], " Feature dimension:", X_val.shape[1])

# Step 3: Standardize the features (important for Logistic Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Step 4: Encode the target labels as integers
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)

# Step 5: Train Logistic Regression model
clf = LogisticRegression(
    solver='lbfgs',            # Optimizer for logistic regression
    max_iter=20000,            # Maximum iterations
    random_state=42
)

clf.fit(X_train_scaled, y_train_encoded)

# Step 6: Prediction and Evaluation
y_pred_encoded = clf.predict(X_val_scaled)

# Convert predictions back to original labels
y_pred = label_encoder.inverse_transform(y_pred_encoded)

# Classification report
print("Classification Report:")
print(classification_report(y_val, y_pred, target_names=['100', '50', '50-100']))

# Confusion matrix
cm = confusion_matrix(y_val, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['100', '50', '50-100'], yticklabels=['100', '50', '50-100'])
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# Save the model and scaler
joblib.dump(clf, '/content/drive/MyDrive/logistic_regression_model.pkl')
joblib.dump(scaler, '/content/drive/MyDrive/scaler.pkl')
joblib.dump(label_encoder, '/content/drive/MyDrive/label_encoder.pkl')
```

```python
import os
import torch
import numpy as np
from PIL import Image
import cv2
from sklearn.preprocessing import StandardScaler
import joblib
from transformers import GroundingDinoProcessor, GroundingDinoForObjectDetection
from ultralytics import SAM
from skimage.feature import hog, local_binary_pattern
from skimage.color import rgb2gray
from sklearn.metrics import classification_report
import warnings

# Suppress the specific warning
warnings.filterwarnings("ignore", message=".*does not have valid feature names.*")


# Load pre-trained models and scaler
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load Grounding DINO
processor = GroundingDinoProcessor.from_pretrained("IDEA-Research/grounding-dino-base")
dino_model = GroundingDinoForObjectDetection.from_pretrained("IDEA-Research/grounding-dino-base")
dino_model.to(device)
dino_model.eval()

# Load SAM model
sam_model = SAM("sam2_l.pt")
sam_model.model.to(device)

# Load the trained LightGBM model
clf = joblib.load('/content/drive/MyDrive/lightgbm_model.pkl')

# Load the scaler used during training
scaler = joblib.load('/content/drive/MyDrive/scaler.pkl')

# === Caption Preprocessing ===
def preprocess_caption(caption: str) -> str:
    result = caption.lower().strip()
    return result if result.endswith(".") else result + "."

# === Feature extraction function ===
def extract_features(image):
    # Resize & convert to grayscale
    img_resized = cv2.resize(image, (128, 128))
    gray_float = rgb2gray(img_resized)
    gray_uint8 = (gray_float * 255).astype(np.uint8)

    # Extract HOG features (multi-scale)
    hog_params = {
        'orientations': 9,
        'pixels_per_cell': (8, 8),
        'cells_per_block': (2, 2),
        'block_norm': 'L2-Hys'
    }
    hog_feat = []
    for scale in [0.5, 1, 1.5]:  # Multi-scale
        scaled_img = cv2.resize(gray_float, (int(128 * scale), int(128 * scale)))
        hog_feat.append(hog(scaled_img, **hog_params))
    hog_feat = np.concatenate(hog_feat)

    # Extract LBP features
    radius = 1
    n_points = 8 * radius
    lbp = local_binary_pattern(gray_uint8, n_points, radius, method='uniform')
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-6)

    # Local brightness features
    local_mean = np.mean(gray_uint8)
    local_variance = np.var(gray_uint8)

    # Concatenate features
    final_feat = np.concatenate([hog_feat, hist, [local_mean, local_variance]])
    return final_feat

# === Image Processing and Prediction Function ===
def predict_image(image_path):
    # Load and process the image
    image = Image.open(image_path).convert("RGB")
    image_rgb = np.array(image)
    H, W = image_rgb.shape[:2]

    # Grounding DINO inference
    inputs = processor(images=image, text=preprocess_caption("human eye"), return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = dino_model(**inputs)

    postprocessed_outputs = processor.image_processor.post_process_object_detection(
        outputs, target_sizes=[(H, W)], threshold=0.4
    )
    results = postprocessed_outputs[0]
    boxes = results["boxes"]
    scores = results["scores"]
    mask = (scores > 0.2) & (scores < 0.8)
    filtered_boxes = boxes[mask].cpu().numpy()

    if len(filtered_boxes) == 0:
        return None

    # SAM segmentation
    sam_results = sam_model(image_path, bboxes=filtered_boxes.tolist(), verbose=False)

    saved_count = 0
    for res in sam_results:
        masks = res.masks.data.cpu().numpy().astype(np.uint8)
        for mask in masks:
            nonzero_pixels = np.count_nonzero(mask)
            pixel_ratio = nonzero_pixels / (H * W)
            if pixel_ratio > 0.95:  # Skip overly large masks
                continue

            mask_3ch = np.repeat(mask[:, :, np.newaxis], 3, axis=2)
            extracted = image_rgb * mask_3ch
            x, y, w, h = cv2.boundingRect(mask)
            cropped = extracted[y:y+h, x:x+w]

            # Feature extraction
            features = extract_features(cropped)
            features_scaled = scaler.transform([features])

            # Prediction with LightGBM
            prediction = clf.predict(features_scaled)

            # Map the predicted category (integer) to a string category using the index
            category = ['100', '50', '50-100'][prediction[0]]
            saved_count += 1

            return category  # Return the first valid prediction

    if saved_count == 0:
        return None
    return None


# === Evaluate the entire test dataset ===
def evaluate_dataset(test_dir):
    all_predictions = []
    all_labels = []

    # Iterate through each category (100, 50, 50-100)
    for category in ['100', '50', '50-100']:
        category_dir = os.path.join(test_dir, category)
        image_files = [f for f in os.listdir(category_dir) if f.lower().endswith(".jpg")]

        # Process each image in the category
        for fname in image_files:
            image_path = os.path.join(category_dir, fname)
            predicted_category = predict_image(image_path)

            # If a valid prediction was made, append it
            if predicted_category:
                all_predictions.append(predicted_category)
                all_labels.append(category)

    # Generate the classification report
    print("Classification Report:")
    print(classification_report(all_labels, all_predictions, target_names=['100', '50', '50-100']))

# === Example Usage ===
test_dir = '/content/spl_data/test'  # Path to the test set folder
evaluate_dataset(test_dir)
```
